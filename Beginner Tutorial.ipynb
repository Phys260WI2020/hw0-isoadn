```python
import matplotlib.pyplot as plt
import numpy as np
plt.ylim(.5,1)
plt.ylabel('Final Grade(G)')
plt.xlabel('Grade on Extra Credit')
plt.title('My grade in Physics 260')

E = np.arange(0.0, 1.0, 0.01)
M1 = .75
M2 = .8
M3 = .85
M4 = .9

G1 = (1-.35*E)*M1+.35*E
G2 = (1-.35*E)*M2+.35*E
G3 = (1-.35*E)*M3+.35*E
G4 = (1-.35*E)*M4+.35*E
G0 = M3+0*E

N1 = ((M3-.35*E)/(1-.35*E))

line1 = plt.plot(E, G1,'r-')
line2 = plt.plot(E, G2,'y-')
line3 = plt.plot(E, G3,'c-')
line4 = plt.plot(E, G4,'m-')
line5 = plt.plot(E, G0,'k-')
line6 = plt.plot(E, N1,'g--')

plt.show()
```
